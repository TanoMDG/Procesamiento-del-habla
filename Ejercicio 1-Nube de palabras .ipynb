```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

Luego de repasar la extraccin de texto de paginas web (webscrapping), utilizar las libreras "requests" y "beatifulsoup" para ello.

```python
#Importo las libreras necesarias
import requests as req
from bs4 import BeautifulSoup as bs

# Direccion de la pgina de donde se extraer el texto
url = "https://es.wikipedia.org/wiki/Inteligencia_artificial"

# Solicitud a la pgina para obtener el contenido
response = req.get(url)

#Con status_code verifico si fue correcta la descarga

if response.status_code == 200:
  print("Descarga correcta")
else:
  print("Error en la descarga")

#Analizo el contenido del HTML con el argumento 'html.parser', que es el analizador de Python para HTML.
#Creando un objeto BeautifulSoup
soup = bs(response.text, "html.parser")

#Ahora se extrae el contenido de todos los prrafos de la pgina.
parrafos = soup.find_all("p")

#Formo un solo string uniendo todos los prrafos, utilizando join()
texto1 = " ".join([p.get_text() for p in parrafos])


#Muestro el texto
texto1

```

```python
# Crear un objeto WordCloud
wordcloud1 = WordCloud(width= 1000, height= 500, background_color="white").generate(texto1)
```

```python
#Para mostrar la imagen que se creo
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud1)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
